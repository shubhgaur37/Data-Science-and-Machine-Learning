```python
#we need 2 arrays x and y which represent the coordinates of the graph
#dimensions of x and y are same
```

```python
#import library(pyplot module of matplotlib library)
import matplotlib.pyplot as plt
```

```python
x=[1,2,3]
y=[2,4,6]
plt.scatter(x,y)#plotting points on a graph in a scattered form
plt.show()#this command is used to show the graph after specifying all
#properties of the graph like labels,axis,scale etc.
```

```python
#making a line graph and connecting the points with a line
x=[1,2,3]
y=[2,4,6]
plt.plot(x,y)
plt.show()
```

```python
#now if we want to plot 2 different lines in the same plot we can also do that
x=[1,2,3]
y=[2,4,6]
x2=[3,4,5]
y2=[2,4,6]
plt.scatter(x,y)
plt.scatter(x2,y2)
plt.plot(x,y)
plt.plot(x2,y2)
plt.show()#we can also set the color for the 2 lines if we want
```

```python
#we can also set the color for the 2 lines if we want by specifying a
#string in plot fxn as an additional parameter,r stands for red,
#g stands for green
x=[1,2,3]
y=[2,4,6]
x2=[3,4,5]
y2=[2,4,6]
plt.scatter(x,y)
plt.scatter(x2,y2)
plt.plot(x,y,'r')
plt.plot(x2,y2,'g')
plt.show()
```

```python
#we can also combine the color parameter with parameters like 
#'-':for line
#'o':for only points,hides the line and shows only points
#'--':for dashed line
#'+': for showing points in the form of +
# '^': points come up as triangles
# 'p': points come in pentagon form
#'-.': LINE SHOWS IN THE FORM OF -.
#we can also set the color for the 2 lines if we want by specifying a
#string in plot fxn as an additional parameter,r stands for red,
#g stands for green
x=[1,2,3]
y=[2,4,6]
x2=[3,4,5]
y2=[2,4,6]
plt.scatter(x,y)
plt.scatter(x2,y2)
plt.plot(x,y,'r-')
plt.plot(x2,y2,'go')
plt.show()
```

## CUSTOMIZING OUR GRAPH

```python
#passing color and markers separately
x=np.arange(1,5,0.2)
y=x**3
plt.plot(x,y,color='green',marker='+')#we can pass color,marker in this
#way also
plt.show()
```

```python
#channging line width
x=np.arange(1,5,0.2)
y=x**3
plt.plot(x,y,color='green',marker='+',linewidth=5)
#we can pass line width also
plt.show()
```

```python
#we can also set labels to our graph
#plt.ylabel:used to set y label
#plt.xlabel:used to set x label
#plt.title: used to set title of graph
#we can also name the plotted line using label parameter in plot fxn
#after setting the plot labels we have to call plt.legend() fxn to show
#them on our graph
```

```python
#plotting x^3 with labels
x=np.arange(0,5,0.2)
y=x**3
plt.plot(x,y,color='cyan',linewidth=3)
plt.xlabel('x')
plt.ylabel('x^3')
plt.title("Graph of x^3")
plt.show()
```

```python
#now suppose we plot 2 different graphs on the same plot,so we need to
#specify for each graph to distinguish bw them
#plotting x^3 and x^2
x=np.arange(0,5,0.2)
y=x**3
y2=x**2
plt.plot(x,y,color='cyan',linewidth=3,label='x^3')
plt.plot(x,y2,color='red',linewidth=3,label='x^2')
#we have specified the labels but still we dont see anything which
#tells us about the two lines,this is because we have to call 
#plt.legend() fxn to show plot labels,then only they will be visible.

plt.xlabel('x')
plt.ylabel('y')
plt.title("x^3 Vs. x^2")
plt.show()
```

```python
x=np.arange(0,5,0.2)
y=x**3
y2=x**2
plt.plot(x,y,color='cyan',linewidth=3,label='x^3')
plt.plot(x,y2,color='red',linewidth=3,label='x^2')

#using plt.legend() fxn
plt.legend()

plt.xlabel('x')
plt.ylabel('y')
plt.title("x^3 Vs. x^2")
plt.show()#labels are showing for both the lines
```

```python
#sometimes we dont want the default axis values,we want to specify our
#axis like from where will they start and till where will they go
#so in these type of cases we use plt.axis fxn
#which takes in a list with 4 parameters
#xstart
#xstop
#ystart
#ystop
x=np.arange(0,5,0.2)
y=x**3
y2=x**2
plt.plot(x,y,color='cyan',linewidth=3,label='x^3')
plt.plot(x,y2,color='red',linewidth=3,label='x^2')

plt.legend()
plt.axis([0,10,0,120])#using plt.axis() to set our own customized axis
plt.grid()#draws a grid for us

plt.xlabel('x')
plt.ylabel('y')
plt.title("x^3 Vs. x^2")
plt.show()
```

```python
#sometimes we want to mark a point and show some information about it
#we can call a fxn called plt.text()
# we have to give it x and y and the text to show at that point
#we can also pass the fontsize in plt.text(optional)
x=np.arange(0,5,0.2)
y=x**3
y2=x**2
plt.plot(x,y,color='cyan',linewidth=3,label='x^3')
plt.plot(x,y2,color='red',linewidth=3,label='x^2')

plt.legend()
plt.axis([0,10,0,100])#using plt.axis() to set our own customized axis
plt.grid()#draws a grid for us
plt.text(1.5,25,'Shubh',fontsize=12)
plt.xlabel('x')
plt.ylabel('y')
plt.title("x^3 Vs. x^2")
plt.show()
```

## BUBBLE CHARTS

```python
#Useful when we want to show multiple types of data in the same graph
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
year=[2012,2013,2014,2015,2016,2017]
salary=[12,13,14,17,19,20]
population=[100,120,180,250,300,370]#population at ith index corresponds
#with salary at ith index.i.e. how many people have salary matching
#salary[i] in population[i]
```

```python
#normal scatter graph
plt.scatter(year,salary)
plt.show()
```

```python
#a bubble graph is same as scatter graph the only difference is that
#points have different sizes,size which represent some information
#we can also set the size of points in a scatter graph by specifying
#s=some value in the plt.scatter fxn.
plt.scatter(year,salary,s=100)#size of all the points increased
plt.show()
```

```python
#now if we want to pass the size for each point we can pass an array
#which contain sizes of all the points to s
#so using this we make a bubble chart
plt.scatter(year,salary,s=population)#size of all the points increased
plt.show()#different sizes which represent population
```

```python
#we can also do the similar thing with color attribute in scatter
#by passing an array to  (c) attribute or just a value
#if we only pass a value then all the points are of the same color
#if we pass an array of values then the points will be of different 
#color
plt.scatter(year,salary,s=population,c='green')
plt.show()
```

```python
#giving title
plt.title('Bubble Chart')
plt.scatter(year,salary,s=population,c='green')
plt.show()
```

```python
#alpha parameter:defines how opaque our points are
#as we decrease alpha they become more and more transparent
#by default our alpha is 1
plt.title('Bubble Chart')
plt.scatter(year,salary,s=population,c='green',alpha=0.5)
plt.show()
```

```python
c=np.arange(len(year))#making an array of colors
```

```python
plt.title('Bubble Chart')
plt.scatter(year,salary,s=population,alpha=0.5,c=c)#passing an array of colors
plt.show()
```

```python
#edge color attribute:adds edgecolor to the points
# by default edge colors are same as point colors
plt.title('Bubble Chart')
plt.scatter(year,salary,s=population,alpha=0.5,c=c,edgecolor='red')#all edge colors are red
plt.show()
```

```python
#change marker: we can use marker parameter to change the marker instead of a circle .

plt.title('Bubble Chart')
plt.scatter(year,salary,s=population,alpha=0.5,c=c,edgecolor='red',marker='^')#changed the marker to *
plt.show()
```

```python
#Now if we want to specify the population along with the points we can call plt.text() fxn with x and y
#and if we want text to appear somewhere on the right then we need to specify appropriate x and y
plt.title('Bubble Chart')
plt.scatter(year,salary,s=population,alpha=0.5,c=c,edgecolor='red',marker='^')#changed the marker to *
n=len(year)
plt.axis([2011,2020,10,25])
for i in range(n):
    plt.text(year[i]+0.2,salary[i],str(population[i]))
plt.show()#showing population against each entry
```

## PIE CHART

```python
#Suppose I have student population of 4 classes,so we can make a pie
#graph which gives me a depiction of how the population is distributed
#among the 4 classes
```

```python
#Example:
#Student population of 4 classes
student=[15,20,9,35]
plt.pie(student)
plt.axis('equal')#makes the graph circle if it is in oval shape by making
#the axes equal
plt.show()#it doesnt look very good,as it doesn't give us any information
#about which color signifies what
```

```python
#we can pass our customized colors also by passing an array which
#contains color names to colors attribute
student=[15,20,9,35]
colors=['blue','purple','red','pink']
plt.pie(student,colors=colors)
plt.axis('equal')
plt.show()
```

```python
#we can also add labels as it is not showing much relevant information
#we can pass an array of labels to labels attribute to add labels
student=[15,20,9,35]
colors=['blue','purple','red','pink']
labels=['Class A','Class B','Class C','Class D']
plt.title('Split among 4 classes')#adding title
plt.pie(student,colors=colors,labels=labels)
plt.axis('equal')
plt.show()
```

```python
#to add some data inside the circle like percentage which gives us 
#a brief about how the data is splitted we can do that also
#there is a parameter called autopct,to which we can pass a fxn which
#determines what values it should show,or a string which tells it what to 
#show eg.%.1f this tells auto pct to show percentage split to one decimal
#point only
student=[15,20,9,35]
colors=['blue','purple','red','pink']
labels=['Class A','Class B','Class C','Class D']
plt.title('Split among 4 classes')
plt.pie(student,colors=colors,labels=labels,autopct='%.2f')
plt.axis('equal')
plt.show()
```

```python
#we can only see numbers inside the graph,
#to show the percentage sign we need to add '%%' after %.2f to show the
#percentage sign
student=[15,20,9,35]
colors=['blue','purple','red','pink']
labels=['Class A','Class B','Class C','Class D']
plt.title('Split among 4 classes')
plt.pie(student,colors=colors,labels=labels,autopct='%.2f%%')#percentage
#sign is showing
#this makes it very clear that we are showing a percentage
plt.axis('equal')
plt.show()
#suppose if we are not sure what to pass to auto pct we can just
#pass a fxn to auto pct and it will decide what to show inside graph
#according to that fxn
```

```python
#Now suppose we want to highlight certain part of data in pie chart
#so what we do is we show that part of data(slice) disjointed from the 
#chart
#we can do this here also using explode property which takes an array
#which tells us how much we want our slice to explode or be disjointed
#default value for each slice is 0
student=[15,20,9,35]
colors=['blue','purple','red','pink']
labels=['Class A','Class B','Class C','Class D']
plt.title('Split among 4 classes')
explode=[0.1,0,0,0]#explode array
plt.pie(student,colors=colors,labels=labels,autopct='%.2f%%',explode=explode)

plt.axis('equal')
plt.show()
```

```python
# A thing to observe here is that the graph is drawn counterclockwise
#now if we want to draw our graph in clockwise then we need to make
#the counterclock parameter to be False
student=[15,20,9,35]
colors=['blue','purple','red','pink']
labels=['Class A','Class B','Class C','Class D']
plt.title('Split among 4 classes')
explode=[0.1,0,0,0]
plt.pie(student,colors=colors,labels=labels,autopct='%.2f%%',explode=explode,counterclock=False)
#counter clock set to False

plt.axis('equal')
plt.show()
#graph is clockwise now
```

```python
#Another thing to notice is that class A starts from 0 degree in the 
#plane(see start of class A) if we want to change that we can also do
#that using startangle parameter,we can set it to any degree
student=[15,20,9,35]
colors=['blue','purple','red','pink']
labels=['Class A','Class B','Class C','Class D']
plt.title('Split among 4 classes')
explode=[0.1,0,0,0]
plt.pie(student,colors=colors,labels=labels,autopct='%.2f%%',explode=explode,counterclock=False,startangle=35)
#specifying start angle

plt.axis('equal')
plt.show()

```

```python
#setting counterclock to true
student=[15,20,9,35]
colors=['blue','purple','red','pink']
labels=['Class A','Class B','Class C','Class D']
plt.title('Split among 4 classes')
explode=[0.1,0,0,0]
plt.pie(student,colors=colors,labels=labels,autopct='%.2f%%',explode=explode,startangle=35)
#specifying start angle and counter clock =true

plt.axis('equal')
plt.show()

```

## Histogram

```python
#suppose we have given a lot of data and we need to see the frequency of
#data like how many times a particular data point is appearing,etc
```

```python
import matplotlib.pyplot as plt
```

```python
a=[1,2,3,4,4,1,3,4,2,1,1]
```

```python
plt.hist(a)
plt.show()#makes a histogram and shows us the frequency of that data
#this is a small part of data so we see different bars
```

```python
#now lets take some big data
b=[1,2,1,3,5,4,5,5,4,3,3,10,7,6,7,8,8,20,22,11,14]
plt.hist(b)
plt.show()#here 1 seems to be present 6 times,but it is not the case 
#that 1 is present 6 times,the histogram is showing us the combined 
#frequencies of numbers
```

```python
#lets distinguish bw those combined frequencies by showing the edges
plt.hist(b,edgecolor='red')
plt.show()#so the first bar shows us the combined frequencies of 1 2and3
#which is 1->2 2->1 3->3=2+1+3=6
#clubbed numbers together for large data
```

```python
#now to understand which numbers are clubbed lets learn xticks
```

```python
plt.hist(b,edgecolor='red')
plt.xticks([1,2,3,4,5])
plt.show()
```

```python
#But I dont want this I want numbers from 1 to 22 to get 
#a better understanding
import numpy as np
xt=np.arange(23)
plt.hist(b,edgecolor='red')
plt.xticks(xt)
plt.show()
```

```python
#seems like first tower interval is from 1 to 3.1
#and for second tower : 3.1-5.2 approx
#So, first tower represents clubbed frequency of everything in between
#and so on for second third and 4th
#By default hist creates 10 towers and we call it bins
#and size of a bin is calculated by:- 
#(maxnumber-minnumber of data)//10
#eg. in our case
#(22-1)/10=2.1
#this means width of each bin is 2.1,i.e. a new bin is created after
#each interval of 2.1
#We can change the number of bins using bins parameter of hist fxn 
plt.hist(b,bins=25,edgecolor='red')
plt.xticks(xt)
plt.show()
#width of bin=(22-1)/25=0.84
```

```python
#now to show a different bar for each entry of array we need to set 
#bin size to 21 because=(22-1)/21=1
plt.hist(b,bins=21,edgecolor='red')
plt.xticks(xt)
plt.show()
#in all the bins it follows the format:-[a,b):- a is included b is not
#except for the last bin,in the last bin all elements are included[a,b]
```

```python
#we can also decide ourself how many bins we want and how we want them to
#be distributed by passing an array to bins parameter
plt.hist(b,bins=[1,3,6,10,15,20],edgecolor='red')#5 towers
plt.show()#we have not specified 22 in bins array so frequency of 22 is 
#not shown
```

```python
#changing range of x and y axis
plt.hist(b,bins=[1,3,6,10,15,20],edgecolor='red')
plt.axis([0,25,0,10])
plt.show()
```

```python
#problem given heights and weights of 20 students.Plot 2 histograms with
#a bin size of 5
import numpy as np
height=np.random.randint(145,190,size=20)
weight=np.random.randint(40,105,size=20)
print((np.max(height)-np.min(height))/5)
print((np.max(weight)-np.min(weight))/5)#bin width
```

```python
#Question 1:Plot line graph b/w no. of job postings with year
#reading data
d={}
with open(r'C:\Users\Shubh Gaur\Desktop\Coding Ninjas Data Science\13. Plotting Graphs\14. Assignment\amazon_jobs_dataset.csv'
          ,encoding='utf8') as file:
    data=csv.DictReader(file)
    for row in data:
        date=row['Posting_date']
        date=int(date[-5:len(date)])#slicing date string and converting to int
        d[date]=d.get(date,0)+1
date_arr=np.array([i for i in d])
#constructing array for no. of job postings
jobs=np.array([d[i] for i in date_arr])
print(d)
print(date_arr)
print(jobs)
```

```python
#ANS
#Plotting part
plt.plot(date_arr,jobs,color='red')
plt.title('Year Wise Job Openings Amazon')
plt.xlabel('Year')
plt.ylabel('No. of Job Postings')
plt.show()
```

```python
#Ques2.:-Plot a bar graph between months vs job openings
def extractMonth(date):
    index=date.find(' ')#finds first occurence of space
    return date[:index]
d={'January':0,'February':0,'March':0,'April':0,'May':0,'June':0,'July':0,
   'August':0,'September':0,'October':0,'November':0,'December':0}
arr=[]
with open(r'C:\Users\Shubh Gaur\Desktop\Coding Ninjas Data Science\13. Plotting Graphs\14. Assignment\amazon_jobs_dataset.csv'
          ,encoding='utf8') as file:
    data=csv.DictReader(file,skipinitialspace=True)
    for row in data:
        date=row['Posting_date']
        date=extractMonth(date)
        d[date]=d.get(date,0)+1
month=np.array([i for i in d.keys()])
jobs=np.array([i for i in d.values()])
        
```

```python
print(d)
print(month)
print(jobs)
```

```python
#plotting graph
#ANS
plt.bar(month,jobs,width=0.6)
plt.title('Jobs Posting Month Wise Amazon')
plt.xlabel('Months')
plt.ylabel('No. of Job Openings')
plt.xticks(rotation=90)

plt.show()
```

```python
#Ques 3.plot pie chart of job openings of various cities of India
cities={}
with open(r'C:\Users\Shubh Gaur\Desktop\Coding Ninjas Data Science\13. Plotting Graphs\14. Assignment\amazon_jobs_dataset.csv'
          ,encoding='utf8') as file:
    data=csv.DictReader(file)
    for row in data:
        location=row['location']
        if location[0:2]=='IN':#for india
            city=location[8:]
            cities[city]=cities.get(city,0)+1
print(cities)
```

```python
city_arr=np.array([i for i in cities])
jobs=np.array([cities[i] for i in city_arr])
print(city_arr)
print(jobs)
```

```python
#plot pie chart
#ANS
plt.pie(jobs,labels=city_arr,autopct='%.2f%%',counterclock=False)
plt.title('Job Openings Distribution City Wise')
plt.show()
```

```python
#pune is not visible clearly so we can explode it
explode=[0,0,0,0,1.5]
plt.pie(jobs,labels=city_arr,autopct='%.2f%%',counterclock=False,explode=explode,radius=1.2)#using radius parameter to increase size
plt.title('Job Openings Distribution City Wise Amazon')
plt.show()
```

```python
#ques4. plot a scatter graph for java openings year wise
def substr(s,string):
    n=len(s)
    m=len(string)
    if n==0:#empty string is always a substring(general case) 
        return True
    if m<n:
        return False
    if string[:n]==s:#BASE CASE
        return True
    return substr(s,string[1:])
years={}
with open(r'C:\Users\Shubh Gaur\Desktop\Coding Ninjas Data Science\13. Plotting Graphs\14. Assignment\amazon_jobs_dataset.csv'
          ,encoding='utf8') as file:
    data=csv.DictReader(file)
    for row in data:
        qualification=row['BASIC QUALIFICATIONS']
        if substr('Java',qualification) or substr('java',qualification):
            date=row['Posting_date']
            year=int(date[-4:])
            years[year]=years.get(year,0)+1

#creating numpy arrays
year_arr=np.array([i for i in years])
jobs=np.array([years[i] for i in year_arr])
print(years)
print(year_arr)
print(jobs)
            


```

```python
#ANS 
#plotting graph
plt.scatter(year_arr,jobs,s=50,marker='*',c='red',edgecolor='blue')
plt.title('JAVA Job openings year wise Amazon')
plt.xlabel('Year')
plt.ylabel('No. of Job Openings')
plt.show()
```
